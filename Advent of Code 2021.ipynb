```python
## Input
file = open("input.txt", "r")
lines = file.read().splitlines()
```

```python
## Day 1

depths = [int(line) for line in lines]

## Part 1

sum_ = 0
for i in range(1, len(depths)):
    if (depths[i] > depths[i - 1]):
        sum_ += 1
        
print(sum_)

## Part 2

prev_depth = sum(depths[0:3])
sum_ = 0

for i in range(3, len(depths)):
    curr_depth = sum(depths[(i - 2):3])
    if (curr_depth > prev_depth):
        sum_ += 1
    prev_depth = curr_depth

print(sum_)
```

```python
## Day 2

commands = [[line.split(' ')[0], int(line.split(' ')[1])] for line in lines]

def find_direction(direction):
    if (direction == 'forward'):
        return (1, 0)
    if (direction == 'up'):
        return (0, -1)
    if (direction == 'down'):
        return (0, 1)
    return (0, 0)

## Part 1

pos = [0, 0]

for command in commands:
    direction = find_direction(command[0])
    pos[0] += direction[0] * command[1]
    pos[1] += direction[1] * command[1]
    
print(pos[0] * pos[1])

## Part 2

pos = [0, 0]
aim = 0

for command in commands:
    direction = find_direction(command[0])
    aim += direction[1] * command[1] 
    pos[0] += direction[0] * command[1]
    pos[1] += direction[0] * command[1] * aim
    
print(pos[0] * pos[1])
```

```python
## Day 3

import functools

reports = [[int(c) for c in line] for line in lines]

COLS = len(reports[0])

def sum_2d_list(list_):
    return functools.reduce(lambda a, b: [c + d for (c, d) in zip(a, b)], list_)

def is_one_most_common(ones, entries):
    return ones > entries / 2

def is_one_equally_common(ones, entries):
    return ones == entries / 2

## Part 1

sums = sum_2d_list(reports)

gamma   = [is_one_most_common(s, len(reports)) for s in sums]
epsilon = [not g for g in gamma]

gamma   = [str(int(g)) for g in gamma]
epsilon = [str(int(e)) for e in epsilon]

print(int(''.join(gamma), 2) * int(''.join(epsilon), 2))

## Part 2

o   = [[c for c in report] for report in reports]
co2 = [[c for c in report] for report in reports]

for i in range(COLS):
    if (len(o) == 1):
        break
    sums = sum_2d_list(o)
    if (is_one_equally_common(sums[i], len(o))):
        num_to_remove = 0
    else:
        num_to_remove = int(not is_one_most_common(sums[i], len(o)))
    for j in range(len(o) - 1, -1, -1):
        if (o[j][i] == num_to_remove):
            del o[j]

for i in range(COLS):
    if (len(co2) == 1):
        break
    sums = sum_2d_list(co2)
    if (is_one_equally_common(sums[i], len(co2))):
        num_to_remove = 1
    else:
        num_to_remove = int(is_one_most_common(sums[i], len(co2)))
    for j in range(len(co2) - 1, -1, -1):
        if (co2[j][i] == num_to_remove):
            del co2[j]

o = [str(c) for c in o[0]]
co2 = [str(c) for c in co2[0]]
            
print(int(''.join(o), 2) * int(''.join(co2), 2))
```

```python
## Day 4

import functools

BOARD_SIZE = 5

numbers = lines[0].split(',')
boards = [[list(filter(lambda x: x != '', lines[j].split(' '))) 
           for j in range(i, i + BOARD_SIZE)] for i in range(2, len(lines), BOARD_SIZE + 1)]

def mark_board(board, number):
    for i in range(BOARD_SIZE):
        for j in range(BOARD_SIZE):
            if (board[i][j] == number):
                board[i][j] = True

def check_board(board):
    winner = [True for i in range(10)]
    for i in range(BOARD_SIZE):
        for j in range(BOARD_SIZE):
            if (board[i][j] != True):
                winner[i] = False
            if (board[j][i] != True):
                winner[i + 5] = False
    return functools.reduce(lambda a, b: a or b, winner)

def sum_board(board):
    sum_ = 0
    for i in range(BOARD_SIZE):
        for j in range(BOARD_SIZE):
            if (board[i][j] != True):
                sum_ += int(board[i][j])
    return sum_

## Part 1

winning_board = None
done = False

for number in numbers:
    for board in boards:
        mark_board(board, number)
    for board in boards:
        if (check_board(board)):
            winning_board = board
            done = True
            break
    if (done):
        break
    
print(int(number) * sum_board(winning_board))

## Part 2

losing_board = None
done = False

for number in numbers:
    for board in boards:
        mark_board(board, number)
    for board in boards:
        if (check_board(board)):
            if (len(boards) > 1):
                boards.remove(board)
            else:
                losing_board = board
                done = True
    if (done):
        break
        
print(int(number) * sum_board(losing_board))
```

```python
## Day 5

segments = [[list(map(int, l.split(','))) for l in line.split(' -> ')] for line in lines]

size = [0, 0]
for segment in segments:
    for i in range(2):
        for j in range(2):
            if (segment[i][j] > size[j]):
                size[j] = segment[i][j]
size[0] += 1
size[1] += 1

def draw_vertical(field, segment):
    for i in range(min(segment[0][1], segment[1][1]), max(segment[0][1], segment[1][1]) + 1):
        field[segment[0][0]][i] += 1
        
def draw_horizontal(field, segment):
    for i in range(min(segment[0][0], segment[1][0]), max(segment[0][0], segment[1][0]) + 1):  
        field[i][segment[0][1]] += 1

## Part 1

field = [[0 for y in range(size[1])] for x in range(size[0])]

for segment in segments:
    if (segment[0][0] == segment[1][0]):
        draw_vertical(field, segment)
    elif (segment[0][1] == segment[1][1]):
        draw_horizontal(field, segment)

sum_ = 0
for i in range(size[0]):
    for j in range(size[1]):
        if (field[i][j] > 1):
            sum_ += 1
print(sum_)

## Part 2

def draw_diagonal(field_, segment):
    if (segment[0][0] < segment[1][0]):
        r1 = range(segment[0][0], segment[1][0] + 1)
    else:
        r1 = range(segment[0][0], segment[1][0] - 1, -1)
    if (segment[0][1] < segment[1][1]):
        r2 = range(segment[0][1], segment[1][1] + 1)
    else:
        r2 = range(segment[0][1], segment[1][1] - 1, -1)
    for i in zip(r1, r2):
        field[i[0]][i[1]] += 1

field = [[0 for y in range(size[1])] for x in range(size[0])]

for segment in segments:
    if (segment[0][0] == segment[1][0]):
        draw_vertical(field, segment)
    elif (segment[0][1] == segment[1][1]):
        draw_horizontal(field, segment)
    else:
        draw_diagonal(field, segment)
        
sum_ = 0
for i in range(size[0]):
    for j in range(size[1]):
        if (field[i][j] > 1):
            sum_ += 1
print(sum_)
```

```python
## Day 9

heightmap = [[int(c) for c in line] for line in lines]

## Part 1

lowest = [[True for c in h] for h in heightmap]
for i in range(len(heightmap)):
    for j in range(len(heightmap[0])):
        if (i != len(heightmap) - 1):
            x_comp = heightmap[i][j] < heightmap[i + 1][j]
            lowest[i][j] = lowest[i][j] and x_comp
            lowest[i + 1][j] = lowest[i + 1][j] and not x_comp
        if (j != len(heightmap[0]) - 1):
            y_comp = heightmap[i][j] < heightmap[i][j + 1]
            lowest[i][j] = lowest[i][j] and y_comp
            lowest[i][j + 1] = lowest[i][j + 1] and not y_comp

sum_ = 0
for i in range(len(lowest)):
    for j in range(len(lowest[0])):
        if (lowest[i][j]):
            sum_ += heightmap[i][j] + 1
print(sum_)

## Part 2

def optional_append(list_, item, visited):
    if (not item in list_ and heightmap[item[0]][item[1]] != 9 and not item in visited):
        list_.append(item)

lowest_coords = []
basin_sizes = []
for i in range(len(lowest)):
    for j in range(len(lowest[0])):
        if (lowest[i][j]):
            lowest_coords.append((i, j))
            basin_sizes.append(0)

visited = []
for i in range(len(lowest_coords)):
    path = [lowest_coords[i]]
    while (len(path) > 0):
        (x, y) = path.pop(0)
        visited.append((x, y))
        basin_sizes[i] += 1
        if (x > 0):
            optional_append(path, (x - 1, y), visited)
        if (x < len(heightmap) - 1):
            optional_append(path, (x + 1, y), visited)
        if (y > 0):
            optional_append(path, (x, y - 1), visited)
        if (y < len(heightmap[0]) - 1):
            optional_append(path, (x, y + 1), visited)
            
basin_sizes.sort()
print(basin_sizes[-1] * basin_sizes[-2] * basin_sizes[-3])
```

```python
## Day 10

syntaxes = list(lines)

def handle_char(char, queue):
    if (char == '('):
        queue.append(')')
    elif (char == '['):
        queue.append(']')
    elif (char == '{'):
        queue.append('}')
    elif (char == '<'):
        queue.append('>')
    elif (char != queue.pop(-1)):
        return False
    return True

## Part 1

def get_char_error_value(char):
    if (char == ')'):
        return 3
    if (char == ']'):
        return 57
    if (char == '}'):
        return 1197
    if (char == '>'):
        return 25137
    return 0
    
sum_ = 0
for i in range(len(syntaxes)):
    queue = []
    for j in range(len(syntaxes[i])):
        success = handle_char(syntaxes[i][j], queue)
        if (not success):
            sum_ += get_char_error_value(syntaxes[i][j])
            break
print(sum_)

## Part 2

def get_char_incomplete_value(char):
    if (char == ')'):
        return 1
    if (char == ']'):
        return 2
    if (char == '}'):
        return 3
    if (char == '>'):
        return 4
    return 0

scores = [0 for i in range(len(syntaxes))]
for i in range(len(syntaxes)):
    queue = []
    for j in range(len(syntaxes[i])):
        success = handle_char(syntaxes[i][j], queue)
        if (not success):
            break
    if (not success):
        continue
    while (len(queue) > 0):
        scores[i] = scores[i] * 5 + get_char_incomplete_value(queue.pop(-1))

scores = list(filter(lambda x: x != 0, scores))
scores.sort()
print(scores[len(scores) // 2])
```

```python
## Day 11

octopuses = [[int(c) for c in line] for line in lines]
    
def handle_neighbour(octopuses, flashing_coords, item):
    octopuses[item[0]][item[1]] += 1
    if (not item in flashing_coords and not item in visited and octopuses[item[0]][item[1]] > 9):
        flashing_coords.append(item)

def handle_flashing(octopuses, sum_):        
    flashing_coords = []
    for i in range(len(octopuses)):
        for j in range(len(octopuses[0])):
            octopuses[i][j] += 1
            if (octopuses[i][j] > 9):
                flashing_coords.append((i, j))
    
    for (x, y) in flashing_coords:
        octopuses[x][y] += 1
        if (x > 0):
            handle_neighbour(octopuses, flashing_coords, (x - 1, y))
            if (y > 0):
                handle_neighbour(octopuses, flashing_coords, (x - 1, y - 1))
            if (y < len(octopuses[0]) - 1):
                handle_neighbour(octopuses, flashing_coords, (x - 1, y + 1))
        if (x < len(octopuses) - 1):
            handle_neighbour(octopuses, flashing_coords, (x + 1, y))
            if (y > 0):
                handle_neighbour(octopuses, flashing_coords, (x + 1, y - 1))
            if (y < len(octopuses[0]) - 1):
                handle_neighbour(octopuses, flashing_coords, (x + 1, y + 1))
        if (y > 0):
            handle_neighbour(octopuses, flashing_coords, (x, y - 1))
        if (y < len(octopuses[0]) - 1):
            handle_neighbour(octopuses, flashing_coords, (x, y + 1))
        
    for i in range(len(octopuses)):
        for j in range(len(octopuses[0])):
            if (octopuses[i][j] > 9):
                octopuses[i][j] = 0
                sum_ += 1
    return sum_
    
## Part 1

octopuses_p1 = [[o for o in octopus] for octopus in octopuses]

sum_ = 0
for i in range(100):
    sum_ = handle_flashing(octopuses_p1, sum_)
print(sum_)

## Part 2

octopuses_p2 = [[o for o in octopus] for octopus in octopuses]

sum_ = 0
while (True):
    handle_flashing(octopuses_p2, sum_)
    sum_ += 1
    
    all_flashing = True
    for i in range(len(octopuses_p2)):
        for j in range(len(octopuses_p2[0])):
            all_flashing = all_flashing and octopuses_p2[i][j] == 0
    if (all_flashing):
        break
print(sum_)
```

```python
## Day 12

tunnels = [[line.split('-')] for line in lines]

def find_all_paths(caves, visited, key, path, count, passed_lowercase_twice = True):
    visited[key] += 1
    path.append(key)
    if (key == 'end'):
        count += 1
    else:
        for cave in caves[key]:
            if (cave.isupper() or (cave.islower() and visited[cave] < 1)):
                count = find_all_paths(caves, visited, cave, path, count, passed_lowercase_twice)
            elif (not passed_lowercase_twice and cave != 'start' and cave != 'end'):
                count = find_all_paths(caves, visited, cave, path, count, True)
    path.pop()
    visited[key] -= 1
    return count

## Part 1

caves = {}
visited = {}
for tunnel in tunnels:
    for cave in tunnel:
        for i in range(2):
            if (not cave[i] in caves.keys()):
                    caves[cave[i]] = []
            caves[cave[i]].append(cave[(i - 1) % 2])
            visited[cave[i]] = 0

paths = find_all_paths(caves, visited, 'start', [], 0)
print(paths)

## Part 2

caves = {}
visited = {}
for tunnel in tunnels:
    for cave in tunnel:
        for i in range(2):
            if (not cave[i] in caves.keys()):
                    caves[cave[i]] = []
            caves[cave[i]].append(cave[(i - 1) % 2])
            visited[cave[i]] = 0

paths = find_all_paths(caves, visited, 'start', [], 0, False)
print(paths)
```

```python
## Day 13

def is_origami_line(line):
    if (len(line) == 0):
        return False
    try:
        int(line[0])
        return True
    except ValueError:
        return False
    
def extract_coords(fold_coords):
    folds = []
    for line in fold_coords:
        coords = ''
        for i in range(len(line) - 1, -1, -1):
            if (line[i] == '='):
                break
            coords += line[i]
        folds.append((line[i -1], int(coords[::-1])))
    return folds

def fold_x(origami, x):
    width = len(origami)
    height = len(origami[0])
    for i in range(x):
        for j in range(height):
            if (x - i - 1 > -1 and x + i + 1 < width):
                origami[x - i - 1][j] = origami[x - i - 1][j] or origami[x + i + 1][j]
    for i in range(width - 1, x - 1, -1):
        del origami[i]
    
def fold_y(origami, y):
    width = len(origami)
    height = len(origami[0])
    for i in range(width):
        for j in range(y):
            if (y - j - 1 > -1 and y + j + 1 < height):
                origami[i][y - j - 1] = origami[i][y - j - 1] or origami[i][y + j + 1]
    for i in range(width - 1, -1, -1):
        for j in range(height - 1, y - 1, -1):
            del origami[i][j]
    
origami_lines = [line for line in lines if (is_origami_line(line))]
fold_coords = [line for line in lines if ('fold' in line)]
origami_coords = [[list(map(int, line.split(',')))][0] for line in origami_lines]
folds = extract_coords(fold_coords)

size = [0, 0]
for coords in origami_coords:
    for i in range(2):
        if (coords[i] > size[i]):
            size[i] = coords[i]
size[0] += 1
size[1] += 1

## Part 1

origami_p1 = [[True if ([x, y] in origami_coords) else False for y in range(size[1])] for x in range(size[0])]

if (folds[0][0] == 'x'):
    fold_x(origami_p1, folds[0][1])
elif (folds[0][0] == 'y'):
    fold_y(origami_p1, folds[0][1])
        
sum_ = 0
for i in range(len(origami_p1)):
    for j in range(len(origami_p1[0])):
        if (origami_p1[i][j]):
            sum_ += 1
print(sum_)

## Part 2

origami_p2 = [[True if ([x, y] in origami_coords) else False for y in range(size[1])] for x in range(size[0])]

for coords in folds:
    if (coords[0] == 'x'):
        fold_x(origami_p2, coords[1])
    elif (coords[0] == 'y'):
        fold_y(origami_p2, coords[1])
        
code = [['#' if o else '.' for o in origami] for origami in origami_p2]
for line in code: #Prints the code, but transposed
    print(line)
```
